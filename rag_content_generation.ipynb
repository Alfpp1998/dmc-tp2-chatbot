```python
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import cm
from reportlab.lib import colors
from reportlab.platypus import (
    SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle,
    PageBreak, HRFlowable
)
from reportlab.lib.enums import TA_CENTER, TA_LEFT, TA_JUSTIFY
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
import os

PAGE_W, PAGE_H = A4

AZUL_OSCURO = colors.HexColor('#1a2e4a')
AZUL_MEDIO  = colors.HexColor('#2c5f8a')
AZUL_CLARO  = colors.HexColor('#d6e8f5')
GRIS_CLARO  = colors.HexColor('#f4f6f8')
GRIS_TEXTO  = colors.HexColor('#444444')
NARANJA     = colors.HexColor('#e07b2a')
VERDE       = colors.HexColor('#2a7d4f')


def build_styles():
    base = getSampleStyleSheet()
    styles = {}

    styles['titulo'] = ParagraphStyle(
        'titulo', fontSize=24, textColor=AZUL_OSCURO,
        spaceAfter=6, spaceBefore=20, alignment=TA_CENTER,
        fontName='Helvetica-Bold', leading=30)

    styles['subtitulo'] = ParagraphStyle(
        'subtitulo', fontSize=14, textColor=AZUL_MEDIO,
        spaceAfter=4, spaceBefore=4, alignment=TA_CENTER,
        fontName='Helvetica', leading=18)

    styles['h1'] = ParagraphStyle(
        'h1', fontSize=16, textColor=AZUL_OSCURO,
        spaceAfter=8, spaceBefore=18, fontName='Helvetica-Bold', leading=20)

    styles['h2'] = ParagraphStyle(
        'h2', fontSize=13, textColor=AZUL_MEDIO,
        spaceAfter=6, spaceBefore=12, fontName='Helvetica-Bold', leading=16)

    styles['h3'] = ParagraphStyle(
        'h3', fontSize=11, textColor=NARANJA,
        spaceAfter=4, spaceBefore=8, fontName='Helvetica-Bold', leading=14)

    styles['cuerpo'] = ParagraphStyle(
        'cuerpo', fontSize=10, textColor=GRIS_TEXTO,
        spaceAfter=6, spaceBefore=2, fontName='Helvetica',
        leading=14, alignment=TA_JUSTIFY)

    styles['bullet'] = ParagraphStyle(
        'bullet', fontSize=10, textColor=GRIS_TEXTO,
        spaceAfter=3, spaceBefore=1, fontName='Helvetica',
        leading=13, leftIndent=14, bulletIndent=4)

    styles['caption'] = ParagraphStyle(
        'caption', fontSize=8, textColor=colors.grey,
        spaceAfter=4, spaceBefore=2, fontName='Helvetica',
        alignment=TA_CENTER, leading=10)

    styles['destacado'] = ParagraphStyle(
        'destacado', fontSize=10, textColor=AZUL_OSCURO,
        spaceAfter=6, spaceBefore=6, fontName='Helvetica-Bold',
        leading=14, leftIndent=10, rightIndent=10,
        borderPad=6, backColor=AZUL_CLARO)

    styles['meta'] = ParagraphStyle(
        'meta', fontSize=9, textColor=colors.grey,
        spaceAfter=2, spaceBefore=0, fontName='Helvetica',
        alignment=TA_CENTER, leading=11)

    return styles


def make_table(data, col_widths=None, header_color=AZUL_OSCURO):
    t = Table(data, colWidths=col_widths)
    style = [
        ('BACKGROUND', (0, 0), (-1, 0), header_color),
        ('TEXTCOLOR',  (0, 0), (-1, 0), colors.white),
        ('FONTNAME',   (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE',   (0, 0), (-1, 0), 9),
        ('ALIGN',      (0, 0), (-1, 0), 'CENTER'),
        ('ROWBACKGROUNDS', (0, 1), (-1, -1), [colors.white, GRIS_CLARO]),
        ('FONTNAME',   (0, 1), (-1, -1), 'Helvetica'),
        ('FONTSIZE',   (0, 1), (-1, -1), 8.5),
        ('ALIGN',      (1, 1), (-1, -1), 'CENTER'),
        ('ALIGN',      (0, 1), (0, -1), 'LEFT'),
        ('GRID',       (0, 0), (-1, -1), 0.4, colors.HexColor('#cccccc')),
        ('VALIGN',     (0, 0), (-1, -1), 'MIDDLE'),
        ('TOPPADDING', (0, 0), (-1, -1), 5),
        ('BOTTOMPADDING', (0, 0), (-1, -1), 5),
    ]
    t.setStyle(TableStyle(style))
    return t


def build_doc(path, story, title='', author='Equipo de Analítica'):
    doc = SimpleDocTemplate(
        path, pagesize=A4,
        leftMargin=2.2*cm, rightMargin=2.2*cm,
        topMargin=2*cm, bottomMargin=2*cm,
        title=title, author=author)
    doc.build(story)
    print(f'  -> Generado: {path}')


def hr(styles):
    return HRFlowable(width='100%', thickness=1, color=AZUL_MEDIO, spaceAfter=8, spaceBefore=4)


def sp(n=1):
    return Spacer(1, n * 0.35 * cm)
```
